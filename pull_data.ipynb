```python
import wandb
api = wandb.Api()
```

```python
group_name = "Sweep delta, seeds, and omit_states_actions with no walls and Q value output (run for longer)."
```

```python
import pandas as pd
runs = api.runs("rldl/icvi", {"$and": [{"group": group_name}]})

def generate_dfs():
    for run in runs:
        history = pd.DataFrame(run.history())
        history["seed"] = run.config["seed"]
        history["omit_states_actions"] = run.config["train_data_args"]["omit_states_actions"]
        history["delta"] = run.config["bellman_delta"]
        yield history

df = pd.concat(list(generate_dfs()))
df.columns
```

```python
def get_df(delta, omit_states_actions):
    return df[(df["delta"] == delta) & (df["omit_states_actions"] == omit_states_actions)]
```

```python
data = get_df(1, 0)
```

```python
import itertools
for (delta, omit_states_actions) in itertools.product(data.delta.unique(), data.omit_states_actions.unique()):
    df = get_df(delta, omit_states_actions)
    df.to_json(f"data/delta_{delta}_omit_states_actions_{omit_states_actions}.json")
```
